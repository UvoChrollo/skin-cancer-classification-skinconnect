```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def reshaper(arr, new_shape):
    array = np.array(arr[:-1])
    label = arr[-1]
    new_data = np.reshape(array, new_shape)
    return new_data, label
```

```python
df_del = pd.read_csv("../input/skin-cancer-mnist-ham10000/hmnist_8_8_RGB.csv")
df_del.shape
```

```python
shape = (8, 8, 3)

x = df_del.iloc[0]
print(len(x))
```

```python
data, label = reshaper(x, shape)
print(data.shape)
print(label)
```

```python
def show_image(arr,lbl):
    fig, ax = plt.subplots()
    ax.imshow(arr)
    ax.set_title(lbl)
    plt.show()
```

```python
show_image(data, label)
```
